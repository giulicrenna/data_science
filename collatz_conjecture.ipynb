```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Change the root number to change the collatz sequence
root = 459
```

```python
def collatz(seed):
  any_list = [seed]
  while any_list[-1] != 1:
    if any_list[-1] % 2 != 0:
      any_list.append(int(any_list[-1]*3+1))
    elif any_list[-1] % 2 == 0:
      any_list.append(int(any_list[-1]/2))
  return any_list
```

```python
sns.set_theme(style="darkgrid")
collatz_numbers = pd.DataFrame(collatz(root))
len(collatz_numbers)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.lineplot(data = collatz_numbers, palette="tab10", linewidth=2.5);
```
